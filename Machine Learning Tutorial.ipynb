```python
scores.mean()
```

```python
def get_score(n_estimatores):
    my_pipeline1 = Pipeline(steps=[('preprocessor',SimpleImputer()),('model',RandomForestRegressor(n_estimators=n_estimatores, random_state=0))])
    scores = -1*cross_val_score(my_pipeline1,X1,y,cv=5,scoring='neg_mean_absolute_error')
    return scores.mean()
```
