```python
# Histograms for benign and maligant tumors

plt.figure(figsize=(10,5))
sns.histplot(cs[cs['Diagnosis']=='B']['Area (mean)'], color='r')
sns.histplot(cs[cs['Diagnosis']=='M']['Area (mean)'], color='g')
plt.show()
```

#Do malignant tumors have higher or lower values for 'Area (mean)' (relative to benign tumors), on average = Yes

#Which tumor type seems to have a larger range of potential values = Malignant

```python
# KDE plots for benign and malignant tumors
plt.figure(figsize=(10,5))
sns.kdeplot(cs[cs['Diagnosis']=='B']['Radius (mean)'], color='r')
sns.kdeplot(cs[cs['Diagnosis']=='M']['Radius (mean)'], color='g')
plt.show()
```
